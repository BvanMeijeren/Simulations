```python
# Chaos game, based on Numberphile video
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame([[0,0], [15,30],[30,0]], columns=['X', 'Y'])

df
```

```python
# Let's store the reference points separately
# These three points are the basis of the triangle
refPoints = pd.DataFrame(data=df)

refPoints
```

```python
import random as rand

def ChaosGame(n):
    df = pd.DataFrame([[0,0], [15,30],[30,0]], columns=['X', 'Y']) # Basic triangle
    current_point = pd.DataFrame([[15,15]], columns = ['X','Y'])  # This acts as the starting position
    for i in range(n):
        rand_dot = rand.randint(0,2)
        new_point = current_point + (df.iloc[rand_dot] - current_point)/2
        df = df.append(new_point, ignore_index = True)
        current_point = new_point
    
    plt.show(df.plot(x= 'X', y = 'Y', kind = 'scatter'))
```

```python
ChaosGame(1000)
```
