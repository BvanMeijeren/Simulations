```python
#First, create a maze generator
# Demands for this generator:
# 1. needs to have entrance and exit
# 2. a path should ways have at least one adjacent path tile

import pandas as pd
import numpy as np
```

```python
def MazeGenerator(width, height):
    maze = pd.DataFrame(np.random.randint(0,2, size = (height,width)))
    return maze
```
